### Standardizing the values

```python
# DAYS_LAST_PHONE_CHANGE column gives number of days before which the customer changed their phone.
#This can be converted in months by dividing by 30.5(Considering avg for 30 and 31 days) for analysis. Negative value should be removed.
current['MON_LAST_PHONE_CHANGE'] = current['DAYS_LAST_PHONE_CHANGE']//-30.5
```

```python
# binning the data

bins = [0,3,6,12,24,60,150]
slots = ['in 3 months', 'in 6 months','in 1 year', 'in 2 years', 'in 5 years', 'more than 5 years']

current['LAST_PHONE_CHANGE'] = pd.cut(current['MON_LAST_PHONE_CHANGE'],bins = bins,labels = slots)
```

```python
#DAYS_BIRTH is the age of customer at the time of application.
#This can be converted in years by dividing by 365.25(Considering leap years).
current['AGE'] = current['DAYS_BIRTH']//-365.25
```

```python
# binning data
current['AGE_GROUP']= pd.cut(current.AGE,bins=np.linspace(20 ,70,num=11))
```

###  Univariate Analysis for categorical variables with respect to TARGET 0 and TARGET 1

```python
def plot_func(col):
    plt.figure(figsize = (15,7))
    plt.subplot(1, 2, 1)
    tplot = target0[col].value_counts().plot.bar()
    for p in tplot.patches:
        tplot.annotate(format(p.get_height(), '.1f'), 
                           (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha = 'center', va = 'center', 
                           xytext = (0, 9), 
                           textcoords = 'offset points')
    plt.title('Distribution of '+ '%s' %col +' for Non-Defaulter', fontsize=14)
    plt.xlabel(col)
    plt.xticks(rotation = 90, fontsize = 12)
    plt.ylabel('Number of cases for Non-Defaulter')
    plt.subplot(1, 2, 2)
    t1plot = target1[col].value_counts().plot.bar()
    for p in t1plot.patches:
        t1plot.annotate(format(p.get_height(), '.1f'), 
                           (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha = 'center', va = 'center', 
                           xytext = (0, 9), 
                           textcoords = 'offset points')
    plt.title('Distribution of '+ '%s' %col +' for Defaulter', fontsize=14)
    plt.xlabel(col)
    plt.xticks(rotation = 90, fontsize = 12)
    plt.ylabel('Number of cases for Defaulter')
    plt.show()
```

## Conclusions

#### 1. Company should do an excessive background check and history check with credit bureau for customers who are Repeaters for loan application and may have been Rejected earlier by some other company. If such customers are approved for the loan without these checks, then they are more likely to default.

#### 2. Customers whose housing type is 'Rented', the company should take their rent into consideration as expense and then calculate income vs expenses analysis before approving their loans.


#### 3. Company should target customers with age group between 30 to 50, alse those whose Income falls into 'Medium' or 'High' Income category. Education factor can also be taken into consideration in this. Customers with Higher Education are more likely to repay the loan assuming they earn well.

#### 4. Company should look for any previous loans that are open, before approving the loan request. Pervious loan annuity may affect the current loan annuity if current loan is approved. Such customers are high risk customers.

#### 5. Company should take Rating from external sources into consideration while making decision. Customers with low rating given by 2 or more sources can prove to be risk to the company as they might default.

#### 6. Credit amount should be decided after considering customer's Income, open loans (if any), expenses, age and count of dependent family members.
